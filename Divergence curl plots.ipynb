$$ \vec f(x,y) = sin(x) \hat x + sin(y) \hat y $$

```python
theta = np.linspace(-4*np.pi, 4*np.pi, 10)
print(theta)
```

```python
r = 2
```

```python
x = r*np.sin(theta)
x
```

```python
y = r*np.cos(theta)
y
```

```python
z = 10*theta
```

```python
fig = plt.figure()
ax = fig.gca(projection = '3d')
ax.set_xlim(10)
ax.set_ylim(10)
ax.set_zlim(10)

x = 0
y = 0
z = 0

u = 7
v = 6
w = 5

ax.quiver(x,y,z, u,v,w, length=0.5)
plt.show()
```
