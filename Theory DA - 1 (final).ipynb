```python
#importing libraries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ODI.csv', low_memory = False)
df
```

```python
#splitting match date into day and year
df[['Day','year']] = df['Match Date'].str.split(',', expand=True)
df['year'] = df['year'].astype(int)
df=df.drop("Match Date",axis="columns")
```

```python
# Q1] Display the histogram that represents year-wise number of matches played by an input country given.

#Taking the input from user for a particular country
ic = input("Enter the Country")  
a=df.loc[(df['Team 1']==ic)|(df['Team 2']==ic)]

#Counting total number of matches played by the input country in the entire year
b=pd.DataFrame(a.year.value_counts())  
b=b.reset_index()
print(b)

#Plotting the histplot
plt.figure(figsize=(10,6))
sns.histplot(b,x='index',y='year',color='orange')

plt.xlabel('Year')
plt.ylabel('Count')

#The following histplot is for England.
```

```python
# Q8] What are the top three countries who won by runs?

a=TD.filter(['Winner','Margin'])
runs= a[a['Margin'].astype(str).str.contains('runs')] #separating runs in the Margin column 
runs[['runs','del']] = runs['Margin'].str.split(' ', expand=True)
runs=runs.drop(['del','Margin'],axis=1)
runs['runs'] = runs['runs'].astype(int)

runs=runs.sort_values('runs',ascending=False) #sorting values in descending order
print(runs.iloc[[0,1,2]]) # Printing the top 3 countries

# Answer] Top 3 countries who won by runs are Pakistan, England, Ireland.
```
