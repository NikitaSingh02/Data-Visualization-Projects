```python
#importing the dataset
import pandas as pd

data = pd.read_csv("University_Placements.csv", low_memory=False)
data.head()
```

```python
# choosing 7 universities from the dataset and dropping the data for the rest
# The selected colleges are: 
# 1) THAKUR COLLEGE OF ENGINEERING AND TECHNOLOGY
# 2) THADOMAL SHAHANI ENGINEERING COLLEGE
# 3) INDIRA COLLEGE OF ENGINEERING AND MANAGEMENT
# 4) DWARKADAS J. SANGHVI COLLEGE OF ENGINEERING
# 5) PUNE INSTITUTE OF COMPUTER TECHNOLOGY
# 6) PILLAI COLLEGE OF ENGINEERING
# 7) MIT COLLEGE OF ENGINEERING

a = ['THAKUR COLLEGE OF ENGINEERING AND TECHNOLOGY',
     'THADOMAL SHAHANI ENGINEERING COLLEGE', 'INDIRA COLLEGE OF ENGINEERING AND MANAGEMENT',
     'DWARKADAS J. SANGHVI COLLEGE OF ENGINEERING', 'PUNE INSTITUTE OF COMPUTER TECHNOLOGY',
     'PILLAI COLLEGE OF ENGINEERING', 'MIT COLLEGE OF ENGINEERING']

data = data[data['college name'].isin(a)]
```

```python
# printing all the college  names

print(data["college name"].unique())
```

```python
# choosing 5 companies from the dataset and dropping the data for the rest
# The selected companies are:
# 1) INFOSYS
# 2) MORGAN STANLEY
# 3) TCS
# 4) CAPGEMINI
# 5) AMDOCS

b = ['INFOSYS', 'MORGAN STANLEY', 'TCS', 'CAPGEMINI', 'AMDOCS']

data = data[data['name of company'].isin(b)]
```

```python
#importing libraries

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,6))
sns.set_style("whitegrid")
sns.boxplot(x = data["name of company"] , y = data["Salary"], hue = data["college name"], data=data, linewidth=1.5)
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x = data["name of company"] , y = data["Salary"], hue = data["college name"], data=data, linewidth=1.5)
```

```python
plt.figure(figsize=(15,7))
sns.histplot(x="bill_depth_mm", y="body_mass_g", hue="species")
```

```python
plt.figure(figsize=(15,10))
sns.distplot(x = data["name of company"] , y = data["Salary"], hue = data["college name"], data=data, linewidth=2.5)
```
