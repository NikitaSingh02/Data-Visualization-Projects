```python
from plotnine import ggplot, aes, geom_line, geom_point, geom_bar, facet_grid, labs, theme_dark, theme_classic
```

```python
#### ggplot - to plot in different way
#### aes - aesthetics
#### geom_line - line plot
#### geom_point point plot
#### geom_bar - bar chart
#### theme_dark - dark theme
```

```python
(
    ggplot(mpg) + 
   aes(x='class', y='hwy') + 
    geom_point()
)
```

```python
(
    ggplot(mpg) + 
   aes(x='class') + 
    geom_bar()
)
```

```python
(
    ggplot(mpg)
    + facet_grid(facets = "year~class")
    + aes(x = 'displ', y = 'hwy')#, color = 'color')
    + labs( #for labeling
        x = "Engine Size",
        y = "Miles per Gallon",
        title = "Miles per Gallon for Each Year and Vehicle Class"
    )
    + geom_point()
   # + theme_dark()
)
```

```python
(
    ggplot(mpg)
    + facet_grid(facets = "year~class")
    + aes(x = 'displ', y = 'hwy')
    + labs( #for labeling
        x = "Engine Size",
        y = "Miles per Gallon",
        title = "Miles per Gallon for Each Year and Vehicle Class"
    )
    + geom_point()
    + theme_dark()
)
```

```python
(
    ggplot(mpg)
    + facet_grid(facets = "year~class")
    + aes(x = 'displ', y = 'hwy')
    + labs( #for labeling
        x = "Engine Size",
        y = "Miles per Gallon",
        color = "Vehicle Class",
        title = "Miles per Gallon for Each Year and Vehicle Class"
    )
    + geom_point()
    + theme_classic()
)
```

```python
(
    ggplot(mpg)
    + aes(x = 'cyl', y = "hwy", color = "class")
    + labs( #for labeling
        x = "Engine Size",
        y = "Miles per Gallon",
        color = "Vehicle Class",
        title = "Miles per Gallon for Each Year and Vehicle Class"
    )
    + geom_point()
    #+ theme_dark()
    + theme_classic() #getting same without it also
)
```
