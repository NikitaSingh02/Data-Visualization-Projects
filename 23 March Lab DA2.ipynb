```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("C:/Users/Nikita/Downloads/insurance.csv")
df
```

```python
from plotnine import ggplot, aes, geom_line, geom_point, geom_bar, facet_grid, labs, theme_dark, theme_classic,  geom_boxplot, geom_histogram, facet_wrap, geom_count
```

```python
# 1. Generate a boxplot for each region BMI values using ggplot.
(
    ggplot(df) + 
    aes(x='region', y = 'bmi', color = 'region') + 
    geom_boxplot(df)
)
```

```python
# 2. Plot facet_grid considering x axis as charges, y axis as bmi and group by children of 0 to 5. Also, assign 
# different colors to smokers.

(
    ggplot(df, aes(x='charges', y='bmi', color = 'smoker'))
    + geom_point()
    + facet_grid('children~')
    + labs(x='Charges', y='BMI')
)
```

```python
# 3. Create subplots contains the information about smokers and non-smokers of both male and female.
(
    ggplot(df, aes(x='charges', color = 'smoker'))
    + geom_histogram()
    + facet_grid('sex ~ region')
    #+ facet_wrap('region')
    + labs(x='Charges', y='BMI')
)
```

```python
# 4. Plot the number of observations from each region using bubble chart.
a = ggplot(df) + geom_count(mapping = aes(x='region', y='sex', color = 'region')) + labs(title = 'Number of Observations in Each Region')
a
```

```python
# 5. Create histogram for charges based on smokers and non-smokers
(
    ggplot(df, aes(x = 'charges', color = 'smoker'))
    + geom_histogram(bins = 15)
  #  + geom_histogram(aes(x = 'charges'))
   # + facet_grid('children~')
    + labs(x='Charges', y='count')
)
```
