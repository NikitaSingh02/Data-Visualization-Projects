```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('University_Placements.csv', low_memory = False)
df
```

```python
grouped_df = df.groupby("name of company")
grouped_lists = grouped_df["college name"].apply(list)
grouped_lists = grouped_lists.reset_index()
print(grouped_lists)
```

```python
grouped_dfs = df.groupby("college name")
grouped_listss = grouped_dfs["name of company"].apply(list)
grouped_listss = grouped_listss.reset_index()
print(grouped_listss)
```

```python
# Choosing 7 universities from the dataset
f = df.loc[df['college name'].isin(['THAKUR COLLEGE OF ENGINEERING AND TECHNOLOGY',
     'THADOMAL SHAHANI ENGINEERING COLLEGE', 'INDIRA COLLEGE OF ENGINEERING AND MANAGEMENT',
     'DWARKADAS J. SANGHVI COLLEGE OF ENGINEERING', 'PUNE INSTITUTE OF COMPUTER TECHNOLOGY',
     'PILLAI COLLEGE OF ENGINEERING', 'MIT COLLEGE OF ENGINEERING'])]
f
```

```python
# Choosing 5 companies from the dataset
fin = f.loc[f['name of company'].isin(['INFOSYS', 'MORGAN STANLEY', 'TCS', 'CAPGEMINI', 'AMDOCS'])]
fin
```

```python
# Reset index
fin.reset_index(inplace = True, drop = True)
fin
```

```python
x = fin['name of company']
y = fin['Salary']
plt.figure(figsize=(12,6))
sns.set_style("whitegrid")
sns.boxplot(x, y, hue = fin["college name"], linewidth=1.5)
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x, y, hue = fin["college name"], data = fin, linewidth=1.5)
```

```python
plt.figure(figsize=(22,10))
sns.catplot(x, y, data = fin, )
```

```python
sns.catplot(x, y, hue="college name",
               data=fin, kind="violin")
```

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(18,10))
sns.barplot(x, y, data=fin, hue="college name")
```

```python
plt.figure(figsize=(18,10))
sns.catplot(x, y, hue="college name", kind="swarm", data=fin)
```
